```python
def search(query):
    boost = {}

    results = index.search(
        query=query,
        filter_dict={},
        boost_dict=boost,
        num_results=10
    )
    return results
```

```python
def build_prompt(query, search_results):
    prompt_template = """
    You're a gym workout assistant. Answer the QUESTION based on the CONTEXT from the Exercises database.
    Use only the facts from the CONTEXT when answering the QUESTION.
    
    QUESTION: {question}
    
    CONTEXT: 
    {context}
    """.strip()

    context = ""
    
    for doc in search_results:
        context = context + f"type_of_activity: {doc['type_of_activity']}\ntype_of_equipment: {doc['type_of_equipment']}\nbody_part: {doc['body_part']}\ntype: {doc['type']}\nmuscle_groups_activated: {doc['muscle_groups_activated']}\ninstruction: {doc['instruction']}\n\n"
    
    prompt = prompt_template.format(question=query, context=context).strip()
    return prompt
```

```python
def llm(prompt):
    response = client.chat.completions.create(
        model='gpt-4o',
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content
```

```python
def rag(query):
    search_results = search(query)
    prompt = build_prompt(query, search_results)
    response = llm(prompt)
    return response
```

```python
def llm_mini(prompt):
    response = client.chat.completions.create(
        model='gpt-4o-mini',
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content
```

```python
def gen_questions(doc):
    prompt = prompt_template.format(**doc)
    response = llm_mini(prompt)
    return response
```

```python
for i in tqdm(range(0, j)):
    name = i

    if name not in documen:             # checks keys
        quest = gen_questions(df.iloc[i])
        documen[name] = quest           # add to dict
```

```python
final_list = []
```

```python
for i, quests in documen.items():
    #print(type(json.loads(quests)))
    for questions in json.loads(quests).values():
        for question in questions:
            final_list.append((i, question))
```
